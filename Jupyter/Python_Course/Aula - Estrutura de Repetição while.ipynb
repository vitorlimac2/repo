# Estrutura de repetição while

- Estruturas de repetição são utilizadas para executar um mesmo trecho de código múltiplas vezes;
- Assim como as estruturas de decisão, a execução do bloco de comando também depende da avaliação de __condições__.
- Estruturas __while__ e __for__.

- __Ideia geral__ da estrutura __while__:
 - __enquanto__ uma condição for verdadeira, eu executo o bloco de código.
 
> while <condição>: <br>
>&emsp;	<bloco de código executado enquanto a condição for verdadeira>

***
## Exemplo 1: progressão aritmética
- Imprimir os $n=10$ primeiros elementos de uma progressão aritmética (PA) com a razão $r$ igual à 2 e o primeiro elemento $a_{1} = 1$. O comportamento geral de uma PA é:

$a_{1} = a_{1}$ <br>
$a_{2} = a_{1} + r$ <br>
$a_{3} = a_{2} + r$ <br>
$a_{4} = a_{3} + r$ <br>
...<br>
$a_{n} = a_{n-1} + r$<br>

- Resolvendo:

$a_{1} = 1$ <br>
$a_{2} = a_{1} + r = 1 + 2 = 3$ <br>
$a_{3} = a_{2} + r = 3 + 2 = 5$ <br>
$a_{4} = a_{3} + r = 5 + 2 = 7$ <br>
...<br>
$a_{10} = a_{9} + r = 17 + 2 = 19$<br>

```python
### Solução sem utilizar estrutura de repetição e o termo geral da PA ;)
a1 = 1
r = 2
print(a1)

a2 = a1 + r
print(a2)

a3 = a2 + r
print(a3)

a4 = a3 + r
print(a4)

a5 = a4 + r
print(a5)

a6 = a5 + r
print(a6)

a7 = a6 + r
print(a7)

a8 = a7 + r
print(a8)

a9 = a8 + r
print(a9)

a10 = a9 + r
print(a10)
```

❔ Quais comandos são repetidos sucessivamente? <br>
❔ Qual é a condição de parada da digitação de comandos acima?
<br>
- Solução para a PA usando estrutura de repetição:

```python
## contador
n = 1

## valor inicial ou inicialização da variável
a = 1

## razão
r = 2

while n <= 10: ## a estrutura while vai executar o bloco enquanto n for menor ou igual a 10      
    print(a)
    a = a + r
    n = n + 1
```

## Exemplo : números divísiveis por 3
- Imprimir todos os números divisíveis por 3 no intervalo de 1 a um número fornecido pelo usuário;

1. Leia o número inteiro fornecido pelo usuário.
2. Teste os números de 1 a x: 
    1. É divisível por 3?
    2. Se sim, imprime.


```python
numero = int(input("Digite o número:")) ## (1)
x = 1

while x <= numero:         ## (2)
    if x % 3 == 0:         ## (2.a)
        print(x)           ## (2.b)
    x = x + 1
```

(1) Escreva um código em Python que leia um número N e imprima todos os números inteiros de N a 1 (em ordem decrescente). 

```python
n = int(input())

while n != 1:
    n = n - 1
    print(n)
```

(2) Escreva um código em Python que leia a altura de 5 pessoas. Este programa deverá calcular e mostrar : <br>
a. A menor altura do grupo;  <br>
b. A maior altura do grupo; <br>

```python
## menor altura
menor_altura = 1000
## maior altura
maior_altura = 0

aux = 1

while aux <= 5:
    altura = float(input("Digite a altura: "))
    
    ## Testar maior e menor altura
    if altura > maior_altura:
        maior_altura = altura
    elif altura < menor_altura:
        menor_altura = altura
    aux = aux + 1

print("A menor altura é %f e a maior altura é %f" % (menor_altura, maior_altura))
```

(3) Escreva um código em Python que leia números inteiros do teclado. O programa deve ler os números até que o usuário digite 0 (zero). No final da execução, exiba a quantidade de números digitados, assim como a soma e a média aritmética

```python
## utilizando palavra reservada break

num_digitados = 0
soma = 0

## enquanto for verdadeiro

while True:
    n = int(input("Digite um número (Ou zero para encerrar): "))
    if n == 0:
        break
    num_digitados = num_digitados + 1
    soma = soma + n

print("Números digitados = %d" % num_digitados)
print("Soma = %d" % soma)

if num_digitados != 0:
    media_aritmetica = soma/num_digitados
    print("Média aritmética = %d" % media_aritmetica)    
```
